```python
print(df.columns)
df['Country'].unique()
df.rename(columns ={'Life expectancy at birth (years)':'Life_exp'}, inplace = True)
```

Has life expectancy increased over time in the six nations?

```python
countries = list(df['Country'].unique())
# countries.remove('Zimbabwe')

def plot_life_exp_change(df):
    for country in countries:
        sns.lineplot(x = df[df['Country'] == country]['Year'], y = df[df['Country'] == country]['Life_exp'], label = country )
    plt.title('Life expectancy change per country ')  
    plt.show()


plot_life_exp_change(df)

# sns.lineplot(x = df[df['Country'] == 'Zimbabwe']['Year'], y = df[df['Country'] == 'Zimbabwe']['Life_exp'], label = 'Zimbabwe')
# plt.title('Life expectancy change in Zimbabwe')
# plt.show()
```

Has GDP increased over time in the six nations?

```python
lower_gdp_countries = ['Chile', 'Mexico', 'Zimbabwe']
higher_gdp_countries = ['China', 'Germany', 'United States of America']
def plot_gdp_change(df, list):
    for country in list:
        sns.lineplot(x = df[df['Country'] == country]['Year'], y = df[df['Country'] == country]['GDP'], label = country )
    plt.title('GDP change per country')  
    plt.show()
    
plot_gdp_change(df, lower_gdp_countries)
plot_gdp_change(df, higher_gdp_countries)
```

Is there a correlation between GDP and life expectancy of a country?
Make a scatterplot per country

```python
def scatterplot(df):
    for country in countries:
        sns.regplot(x = df[df['Country'] == country]['Life_exp'], y = df[df['Country'] == country]['GDP'])
        plt.title(f'Relation between Life Expectancy and GDP in {country}')
        plt.show()

scatterplot(df)
```
